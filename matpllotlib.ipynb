```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 	Plot a line chart of Product vs Sales.
# 	Add labels and title.
# 	Customize line color or marker.

data = { 'Product': ['Soap', 'Shampoo', 'Toothpaste', 'Lotion', 'Facewash'],
    'Sales': [120, 80, 150, 60, 90],
    'Ratings': [3.5, 4.0, 4.5, 3.0, 4.2],
    'Price': [40, 120, 60, 150, 90]}

df = pd.DataFrame(data)
df
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.plot(df['Product'], df['Sales'], color='red', marker='o')
plt.show()
```

```python
# 	Create a 1x2 subplot:
# o	Left: Bar chart of Product vs Sales
# o	Right: Bar chart of Product vs Ratings
# 	Add titles to each subplot.
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.bar(df['Product'], df['Sales'], color='red')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales')

plt.subplot(1,2,2)
plt.bar(df['Product'], df['Ratings'])
plt.title('Product Ratings')
plt.show()

```

```python
# 	Create a 2x1 figure:
# o	Top plot: Line plot of Product vs Price
# o	Bottom plot: Bar chart of Product vs Sales
# 	Use fig, axes = plt.subplots(...) and set figure size.

fig, axes = plt.subplots(nrows=2, ncols=1,  figsize=(10,10))
axes[0].plot(df['Product'], df['Price'], marker='o')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Price')
axes[0].set_title('Product Price')


axes[1].bar(df['Product'], df['Sales'])
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Sales')
axes[1].set_title('Product Sales')
plt.show()
```

```python
# 	Plot a pie chart of total sales distribution for each product.
# 	Display percentages using autopct='%1.1f%%'
# 	Add title and make one slice "explode" (highlight).

plt.figure(figsize=(8,8))
explode=[0,0,0.1,0,0]
plt.pie(df['Sales'], labels=df['Product'],autopct='%1.1f%%', explode=explode, shadow=True)
plt.title('total sales Distribution')
plt.show()
```

```python
# 	Create a histogram of Ratings.
# 	Use bins=5 and choose a color.
# 	Add a title and axis labels.

plt.hist(df['Ratings'], bins=5, color='cyan')
plt.title('Distribution of Product Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.6)
plt.show()
```
