```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
  # Create Sales Data- Use np.array()
  # to create a 2D array 'sales' with shape (4, 6).
  # - Each row represents a product (A, B, C, D).
  # - Each column represents a month (Jan to Jun).
```

```python
sales = np.array([
 [200, 220, 250, 275, 300, 310], # Product A
 [150, 160, 170, 180, 190, 200], # Product B
 [100, 120, 130, 140, 150, 160], # Product C
 [90, 95, 100, 105, 110, 115] # Product D
])
products = ['A', 'B', 'C', 'D']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
print(sales.shape)
print(sales)
```

```python
# 2. Monthly & Product Analysis
# - Calculate total sales per product (row-wise sum).
# - Calculate total sales per month (column-wise sum).
# - Identify the best-selling product (based on total).
# - Identify the month with the highest overall sales.
```

```python
total_sales_per_product = np.sum(sales, axis=1)
print("Total sales per product:", total_sales_per_product)

total_sales_per_month = np.sum(sales, axis=0)
print("Total sales per month:", total_sales_per_month)

best_selling_product = products[np.argmax(total_sales_per_product)]
print("Best-selling product:", best_selling_product)

highest_sales_month = months[np.argmax(total_sales_per_month)]
print("Month with highest sales:", highest_sales_month)
```

```python
# 3. Percentage Growth
# - Calculate percentage growth from Jan to Jun for each product:
# growth = ((sales[:, 5] - sales[:, 0]) / sales[:, 0]) * 100
```

```python
growth = ((sales[:, 5] - sales[:, 0]) / sales[:, 0]) * 100
print("Percentage growth from Jan to Jun for each product:", growth)
```

```python
# A. Line Plot
# - Plot line graphs for all 4 products.
# - Each line represents one products sales over 6 months.
# - Use different colors and markers.
# - Add legend, title, x/y labels.
```

```python
plt.figure(figsize=(10, 6))

for i in range(len(products)):
    plt.plot(months, sales[i, :], marker='o', label=products[i])

plt.title('Product Sales Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Product')
plt.grid(True)
plt.show()
```

```python
# B. Bar Chart
# - Plot a grouped bar chart showing each product's sales in June.
```

```python
june_sales = sales[:, 5]
bars =plt.bar(products, june_sales, color=['blue', 'green', 'orange', 'purple'])
plt.title("June Sales by Product")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.bar_label(bars)
plt.show()
```

```python
# C. Pie Chart
# - Show the contribution of each product to total sales in June using a pie chart.
```

```python
plt.pie(june_sales, labels=products, autopct='%1.1f%%', startangle=140)
plt.title("June Sales Contribution by Product")
plt.axis('equal')
plt.show()
```
